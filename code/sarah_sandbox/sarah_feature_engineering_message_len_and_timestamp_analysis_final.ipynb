```python
def load_json_file(path):
    with open(path) as f:
        data = json.load(f)
    return data

def make_submission_csv(predictions, ids, submission_name, path = 'C:/Users/Sarah/Documents/GitHub/brokenBNZ/predictions'):
    with open(path+'/'+submission_name+'.csv', 'w') as csvfile:
        field_names = ['request_id', 'requester_received_pizza']
        writer = csv.DictWriter(csvfile, fieldnames = field_names)
        writer.writeheader()
        csv_data = zip(ids, predictions)
        for row in csv_data:
            writer.writerow({field_names[0]:row[0], field_names[1]:int(row[1])})


# Uploading the raw json data.
all_train_data = load_json_file('C:/Users/Sarah/Anaconda/Scripts/train.json')
test_data =  load_json_file('C:/Users/Sarah/Anaconda/Scripts/test.json')

rand.shuffle(all_train_data)

all_train_data_df = pd.DataFrame(all_train_data)
test_df = pd.DataFrame(test_data)

# Creating the training labels:
all_train_labels = np.array(all_train_data_df["requester_received_pizza"])
```

```python

# First test to see if length of message influenced if message was awarded a pizza.

# Creating our new array to measure length.
X_train_length = np.array([len(x['request_text_edit_aware']) for x in all_train_data])
X_test_length = np.array([len(x['request_text_edit_aware']) for x in test_data])
print X_train_length
print X_test_length

pd.DataFrame(X_train_length).describe()
pd.DataFrame(X_test_length).describe()

# Looking at the message length for people who earned a pizza.
# I was going to run this same thing for people who received pizza in the test data, but it does not have a column 'requester_received_pizza' to compare to.
winners_message_len_train = []
for x in all_train_data:
    if x['requester_received_pizza'] == 1:
        text = x['request_text_edit_aware']
        text_len = len(text)
        winners_message_len_train.append(text_len)

    
print winners_message_len_train


# Plotting our winner's message length to see if there is a given cluster, ie. ideal length.
import matplotlib.pyplot as plt

# This tells matplotlib not to try opening a new window for each plot.
%matplotlib inline

plt.title("Winner's Length of Submitted Message \n")
plt.hist(winners_message_len_train, bins = 50)
plt.xlabel("Message Length")
plt.ylabel("Frequency")
plt.show()


plt.title("\nTraining Data - Length of Submitted Message \n")
plt.hist(X_train_length, bins = 50)
plt.xlabel("Message Length")
plt.ylabel("Frequency")
plt.show()

plt.title("\nTest Data - Length of Submitted Message \n")
plt.hist(X_test_length, bins = 50)
plt.xlabel("Message Length")
plt.ylabel("Frequency")
plt.show()


```
